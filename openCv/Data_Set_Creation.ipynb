```python
import numpy as np
import cv2
```

```python
dataset = []

def draw(event, x,y, flag, param):
    if event == cv2.EVENT_FLAG_LBUTTON:
        cv2.circle(img, (x,y),3,(255,0,0),2)
        dataset.append((x,y,0))
    elif event == cv2.EVENT_FLAG_RBUTTON:
        cv2.circle(img, (x,y),3,(0,255,0),2)
        dataset.append((x,y,1))

img = np.zeros((500,500,3), np.uint8)
cv2.namedWindow("Window_nor", cv2.WINDOW_NORMAL)
cv2.setMouseCallback("Window_nor", draw)

while True:
    cv2.imshow("Window_nor", img)
    if cv2.waitKey(1) & 255 == ord("q"):
        break 
cv2.destroyAllWindows()
```
